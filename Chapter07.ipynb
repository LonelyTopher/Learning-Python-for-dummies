```python
def Hello3(Greeting = "No Value Supplied"):
  print(Greeting)

Hello3()
Hello3("This is a string")
```

```python
def Hello4(ArgCount, *VarArgs):
  print("you passed ", ArgCount, " Arguments.")
  for Arg in VarArgs:
    print(Arg)

Hello4(3, "One", "Two", "Three")
```
