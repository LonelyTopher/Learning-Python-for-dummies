# Working with Loops

###Simple for loops

```python
LetterNum = 1
for Letter in "Howdy!":
  print("Letter ", LetterNum, " is ", Letter)
  LetterNum += 1
```

```python
Value = input("Type less than 6 characters: ")
LetterNum = 1
for Letter in Value:
  print("Letter", LetterNum, " is ", Letter)
  LetterNum += 1
  if LetterNum > 6:
    print("The string is too long!")
    break
```

```python
LettNum = 1
for Letter in "Howdy!":
  if Letter == "w":
    continue
    print("Encountered W, letter not processed")
  print("Letter ", LettNum, " is ", Letter)
  LettNum += 1
```

```python
LetterNum = 1
for Letter in "Howdy!":
  if Letter == "w":
    pass # add some code here later to process w #
  print("Letter ", LetterNum, " is ", Letter)
  LetterNum += 1
```

```python
Value = input("Type less than 6 characters: ")
LetterNum = 1
for Letter in Value:
  print("Letter", LetterNum, " is ", LetterNum)
  LetterNum += 1
  if (Letter == "" or Letter == Value[-1]):
    break
else:
  print("The string is blank.")
```

###While Loops

```python
Sum = 0
while Sum < 5:
  print(Sum)
  Sum += 1
```

### Nesting Loop Statements

```python
x = 1
y = 1
print('{:>4}'.format(' '), end= ' ')
for x in range(1, 11):
  print('{:>4}'.format(x), end= ' ')
print()
for x in range(1, 11):
  print('{:>4}'.format(x), end= ' ')
  while y <= 10:
    print('{:>4}'.format(x * y), end= ' ')
    y += 1
  print()
  y = 1
```
